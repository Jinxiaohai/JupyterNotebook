```python
import matplotlib.pyplot as plt
import numpy as np
"""
author : xiaohai
email : xiaohaijin@outlook.com
"""

x = np.linspace(0, 10, 1000)
y = np.sin(x)
z = np.cos(x**2)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label="$sin(x)$", color='red', linewidth=2)
plt.plot(x, z, label="$cos(x^2)$")

plt.xlabel('Time(s)')
plt.ylabel('Volt')
plt.title("example")
plt.ylim(-1.2, 1.2)
plt.legend()

plt.show()
```
