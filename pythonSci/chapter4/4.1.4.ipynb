## 绘制多个子图 subplot()函数

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 4), dpi=100)
for idx, color in enumerate('rbcky'):
    plt.subplot(3, 2, 1+idx, facecolor=color)
plt.show()
```

```python
import numpy as np
plt.figure(1)
plt.figure(2)
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)

x = np.linspace(0, 3, 1000)
for i in range(5):
    plt.figure(1)
    plt.plot(x, np.exp(i*x/3))
    plt.sca(ax1)
    plt.plot(x, np.sin(i*x))
    plt.sca(ax2)
    plt.plot(x, np.cos(i*x))

plt.show()
```

## 我们可以调用plt.subplots()函数一次绘制多个子图

```python
fig2, ax2 = plt.subplots(3, 3)
plt.show()
```

## 我们还可以更加复杂的绘制子图subplot2grid()
> 其函数形式为subplot2grid(shape, loc, rowspan=1, colspan=1, **kwargs)

```python
fig3 = plt.figure(figsize=(6, 6), dpi=120)
plt.subplot2grid((3, 3), (0, 0), colspan=2)
plt.subplot2grid((3, 3), (1, 2), rowspan=2)
plt.show()
```
